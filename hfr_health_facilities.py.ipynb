```python
import pymysql
import pandas as pd

# create a connection object
connection = pymysql.connect(
    host='localhost',
    user='root',
    password='mysql',
    db='altschool_db'
)

# create a cursor object
cursor = connection.cursor()

# Create the table in the database
create_table_query = '''
CREATE TABLE IF NOT EXISTS NHFR (
    id INT PRIMARY KEY,
    Start_Date VARCHAR(255),
    Facility_Name VARCHAR(255),
    State VARCHAR(255),
    LGA VARCHAR(255),
    Ward VARCHAR(255),
    Ownership VARCHAR(255),
    Ownership_Type VARCHAR(255),
    Facility_Level VARCHAR(255),
    Facility_Level_Option VARCHAR(255),
    Longitude VARCHAR(255),
    Latitude VARCHAR(255),
    Operational_Days VARCHAR(255),
    Operation_Hour VARCHAR(255),
    Operation_Status VARCHAR(255),
    Registration_Status VARCHAR(255),
    License_Status VARCHAR(255),
    No_Of_Doctors VARCHAR(255),
    No_Of_Pharmacists VARCHAR(255),
    Dentist VARCHAR(255),
    Pharmacy_Technicians VARCHAR(255),
    No_Of_Nurses VARCHAR(255),
    No_Of_Lab_Scientists VARCHAR(255),
    Midwifes VARCHAR(255),
    Lab_Technicians VARCHAR(255),
    Nurse_Midwife VARCHAR(255),
    HIM_Officers VARCHAR(255),
    Community_Health_Officer VARCHAR(255),
    Community_Extension_Workers VARCHAR(255),
    Jun_Community_Extension_Worker VARCHAR(255),
    Dental_Technicians VARCHAR(255),
    Env_Health_Officers VARCHAR(255),
    No_Of_Beds VARCHAR(255),
    No_Of_Attendants VARCHAR(255),
    Region VARCHAR(255),
    Start_Year VARCHAR(255),
    Start_Month VARCHAR(255),
    No_Of_Health_Personnel VARCHAR(255)
);
'''

cursor.execute(create_table_query)



for index, row in df.iterrows():
    insert_query = f"""
        INSERT INTO NHFR (
            id, Start_Date, Facility_Name,
            State, LGA, Ward, Ownership,
            Ownership_Type, Facility_Level, Facility_Level_Option,
            Longitude, Latitude,
            Operational_Days,
            Operation_Hour, Operation_Status, Registration_Status,
            License_Status, No_Of_Doctors, No_Of_Pharmacists, Dentist,
            Pharmacy_Technicians, No_Of_Nurses, No_Of_Lab_Scientists,
            Midwifes, Lab_Technicians, Nurse_Midwife, HIM_Officers,
            Community_Health_Officer, Community_Extension_Workers,
            Jun_Community_Extension_Worker, Dental_Technicians,
            Env_Health_Officers, No_Of_Beds,
            No_Of_Attendants,
            Region, Start_Year, Start_Month,
            No_Of_Health_Personnel
        ) 
        
        VALUES (
            '{row['id']}','{row['Start_Date']}', '{row['Facility_Name']}',
            '{row['State']}', '{row['LGA']}', '{row['Ward']}', '{row['Ownership']}',
            '{row['Ownership_Type']}', '{row['Facility_Level']}', '{row['Facility_Level_Option']}',
            '{row['Longitude']}', '{row['Latitude']}',
            '{row['Operational_Days']}',
            '{row['Operation_Hour']}', '{row['Operation_Status']}', '{row['Registration_Status']}',
            '{row['License_Status']}', '{row['No_Of_Doctors']}', '{row['No_Of_Pharmacists']}', '{row['Dentist']}',
            '{row['Pharmacy_Technicians']}', '{row['No_Of_Nurses']}', '{row['No_Of_Lab_Scientists']}',
            '{row['Midwifes']}', '{row['Lab_Technicians']}', '{row['Nurse_Midwife']}', '{row['HIM_Officers']}',
            '{row['Community_Health_Officer']}', '{row['Community_Extension_Workers']}',
            '{row['Jun_Community_Extension_Worker']}', '{row['Dental_Technicians']}',
            '{row['Env_Health_Officers']}', '{row['No_Of_Beds']}',
            '{row['No_Of_Attendants']}', '{row['Region']}', '{row['Start_Year']}', '{row['Start_Month']}','{row['No_Of_Health_Personnel']}'
            )"""

    cursor.execute(insert_query)
        
    # Commit the changes
    connection.commit()
```
